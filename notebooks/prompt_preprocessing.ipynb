```python
import pandas as pd


prompts_df = pd.read_csv('../assets/prompts_trial_2.csv', header=None, names=['input', 'unused1', 'unused2', 'edit', 'output'])
prompts_df.head()
```

```python
# split 'input' by '#', and make the first two column 'input' and 'input_'
prompts_df[['input', 'input_']] = prompts_df['input'].str.split('#', expand=True).iloc[:, :2]
# split 'edited' by '#', and make the first two column 'output' and 'output_'
prompts_df[['output', 'output_']] = prompts_df['output'].str.split('#', expand=True).iloc[:, :2]
prompts_df.head()
```

```python
prompts_df[['input', 'edit', 'output']].to_json('../assets/prompts_trial_2.jsonl', orient='records', lines=True)
```
